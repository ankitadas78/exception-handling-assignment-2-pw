ques1_explain why we use the exception class while creating a custom exception

```python
#customs exceptions enhance the usanility of a class , an exception class should have init and str magic methods which are called durin the exception handling process , whenever we want to create a succesfull program we want to use the custom exception

#example

class validprice(Exception):
    
    def __init__(self , amount):
        self.amount = amount
```

```python
def price_details(price):
    if price < 200 :
        raise validprice("no i do not buy in this amount because of quality issue")
    elif price > 1000 :
        raise validprice("I can't buy in this amount because it is very high amount")
    else :
        print("yes I can buy this product in this amount")
```

```python
try :
    price = int(input("enter your price rate"))
    price_details(price)
except validprice as e :
    print(e)
```

ques6_list down some best practices for exception handling in python

```python
#use always a specific exception

try :
    10/0
except Exception as e :
    print(e)
```

```python
#print always a proper message

try :
    10/0
except ZeroDivisionError as e :
    print("i am trying to handle a zerodivisoin error" , e)
```

```python
#always try to log your error

import logging
logging.basicConfig(filename = "error.log" ,level = logging.ERROR)
try :
    10/0
except ZeroDivisionError as e:
    logging.error("i am trying to handle a zerodivsion error {} ".format(e))
```

```python
#always avoid to write a multiple exception handling
try :
    10/0
except FileNotFoundError as e :
    logging.error("i am handling file not found error {} ".format(e))
except AttributeError as e :
    logging.error("i am trying to handle a attribute error {} ".format(e))
except ZeroDivisionError as e:
    logging.error("i am trying to handle a zerodivsion error {} ".format(e))
```
